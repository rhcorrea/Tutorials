Quando usei a função isnull(), perguntei ao Python se aquele valor era nulo. Se o valor é nulo, a resposta é True; caso contrário, a resposta é False.

#### E agora? que fazer com esses 19.620 valores faltantes?

Bom, temos algumas alternativas:

A primeira delas é simplesmente descartar todas as linhas que possuam o valor NaN (Not a Number).

A segunda é "completar" os dados usando alguma técnica. Isso é chamado imputação (imputation).

#### Como decidir o que fazer? Largar os dados ou fazer imputation?

Aqui eu seguirei com imputation para demonstrar como pode ser feito.

#### Os comandos são: criar BoardImpute igual a df; depois, completar os NaNs com a mediana (fillna), para todas as observações.

```python
BoardImpute.isnull().sum()
```

### 2.1. Olhando um pouco mais a fundo as correlações.

Já que encontramos uma tendência visual para a assertiva: músicos famosos criar hits famosos, vamos tentar trazer um pouco mais de "tutano" para a análise.

Qual o tamanho da correlação que temos entre sucesso do artista versus a música?

```python
#Aqui sempre é um comando para ajustar o tamanho do gráfico: 20 x 10.
plt.figure( figsize = (20,10))
#criando um objeto que calcula a correlação das variáveis, para guardar e usar como argumento do gráfico.
corr = BoardImpute.corr()
#Aqui é o mapping, para fazer a paleta de cores.
cmap = sns.diverging_palette(220, 10, as_cmap = True)
#comando no seaborn para criar um mapa de calor, usando o objeto corr (criado anteriormente)
sns.heatmap(corr, cmap = cmap,  square = True, linewidth = .5, annot = True)
# o comando annot = True é para apresentar os valores de correlação dentro dos quadrados.
```

Daqui podemos ver que realmente, apenas a popularidade do Artista no Spotify versus a Popularidade da Música no Spotify se relacionam, com um segundo lugar para o ano da música.
